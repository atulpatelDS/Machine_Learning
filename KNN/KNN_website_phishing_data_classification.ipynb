## Phishing Site Detection

https://data.mendeley.com/datasets/h3cgnj8hft/1

Description of this data

This dataset contains 48 features extracted from 5000 phishing webpages and 5000 legitimate webpages, which were downloaded from January to May 2015 and from May to June 2017. 
    
    The  data coding is as follows:
    "1"  - Legitimate Webpage
    "0"  - Phishing Webpage

```python
df_phishing["CLASS_LABEL"] = pd.to_numeric(df_phishing["CLASS_LABEL"])
```

```python
### do level encoding on "CLASS_LABEL"
from sklearn.preprocessing import LabelEncoder,StandardScaler
```

```python
col_name = list(df_phishing)
col_name
```

```python
y = df_phishing["CLASS_LABEL"]
```

```python
col_name.remove("CLASS_LABEL")
```

```python
col_name
```

```python
X = df_phishing[col_name]
```

```python
X.head()
```

```python
y.head()
```

```python
print(X.shape)
print(y.shape)
print(type(X))
print(type(y))
```

```python
classifier_knn = KNeighborsClassifier()
classifier_knn.fit(X,y)
```

```python
# Computing the Cross validation score with 5-fold cross validation
cv_score_knn = cross_val_score(classifier_knn, X, y, cv=5)
print("Cross Validation score : " + str(cv_score_knn))
print("Cross Validation Mean score : " + str(cv_score_knn.mean()))
```

```python
### Find the best score and parameter and estimater using Grid Search
from sklearn.model_selection import GridSearchCV
```

```python
### We now use a grid search to find the optimal.
parameter = {"n_neighbors" : range(1,11),"weights":["uniform","distance"],"metric":["euclidean", "manhattan", "chebyshev", "minkowski"]}
```

```python
cv_gs_knn = GridSearchCV(classifier_knn,param_grid=parameter,cv= 10)
```

```python
cv_gs_knn.fit(X,y)
```

```python
print("best estimator :",cv_gs_knn.best_estimator_)
print("best parameter :",cv_gs_knn.best_params_)
print("best score :",cv_gs_knn.best_score_)
```
